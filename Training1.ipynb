#Mutable and immutable datatypes in python
Mutable types are those whose value can be changed after initializing - **List, Set, Dictionaries**
Immutable sre those whose value cant be changed -**String, Numbers, Tuples**

```python
var1 =1;var2=3.5;var3 = 5+7j;
print(var1,var2,var3);print(type(var1),type(var2),type(var3));
```
