## **Measures Of Dispersion**
It represents how spread out the data is.



1.   **Range** -  It is the distance between the    maximum and minimum values.  Highly sensitive to outliers. A single extreme value can make the dataset look much more spread out than it actually is.            
Formula: Range = Max — Min.



2.  **Interquartile Range (IQR)** - is another measure of spread, but unlike variance, it is not affected much by outliers.
The IQR is simply the distance between Q1 and Q3.

    *Formula* = **IQR = Q3-Q1** - This tells you the middle 50% of the scores lie between which range. It is partially resilliant to anomalies.

    *To check if a data is an outlier* :-

    Max Oulier =  Q3 + 1.5 * IQR

    Min Oulier =  Q1 - 1.5 * IQR

Real world scenario : Suppose salaries are:
40k,42k,43k,44k,45k,46k,2,000k.

The average and variance become very large because of the CEO's salary.But the IQR still tells you:

"Most employees earn between about 42k and 46k."

**IQR asks: "How spread out are the middle 50 students?"**

3.  **Variance (sigma² or s²)** - The mean tells you the center of the data, but it doesn't tell you how spread out the data is. That's why we calculate variance.
**Variance measures how much the values deviate from the mean.**

    ex: for there are two datasets:

    Server A : 100, 101, 99, 100, 100

    Server B : 20, 50, 100, 150, 180

    **For both mean is same - 100 ms**

    But they are clearly very different.

    Server A variance = 0.4 ms²

    Server B variance = 3560 ms²

    So Variance (or, more commonly, standard deviation) tells you that:

    **Server A is very consistent**.

    **Server B is highly unpredictable.**
     
     In such scenarios we calculate Variance measures variance, which means how much the values deviate from the mean.

     **Variance asks: "How far is every student's score from the average?"**

4.  **Standard Deviation (sigma or s)**

    Variance = Average of the squared differences from the mean.

    Standard Deviation = Square root of the variance.

   ** Significance: It brings the measure back to the original units of the data.**
	​
    For the bove Server A and B sample, Standard deviation is

    Server A  = √0.4 ≈ 0.63 ms

    Server B = √3560 ≈ 59.67 ms

    Interpretation:

**Server A**

A standard deviation of 0.63 ms means:

Most response times are within about 0.63 ms of the average (100 ms).

Very consistent.

**Server B**

A standard deviation of 59.67 ms means:

Response times typically vary by about 60 ms from the average.

Very inconsistent.

5.   **Coefficient of Variation (CV)**

This is a “relative” measure of dispersion, expressed as a percentage. It is useful when comparing the spread of two datasets that have different units or wildly different means..

The coefficient of variation (CV) is calculated using the formula:

**CV = σ / μ** , where σ is the standard deviation and μ is the mean

It is usually expressed as a percentage:

**CV=    (Mean / Standard Deviation)×100%**

Why do we need CV?

Suppose you're comparing two servers.

**Server A**
Mean response time = 100 ms
Standard deviation = 10 ms

**CV: (100/10)×100=  10%**

**Server B**
Mean response time = 1000 ms
Standard deviation = 50 ms

**CV: (1000/50)×100=  5%**


Which server is more consistent?

If you only looked at the standard deviation:

Server A: 10 ms

Server B: 50 ms

**You might think Server B is worse.**

But relative to their averages:

**Server	   Mean	     SD	     CV**

**A**	      100 ms	  10 ms	   10%

**B**	      1000 ms	  50 ms	   5%

Server B actually has less relative variation. Its response times vary by only 5% of its average, compared to 10% for Server A.

That's exactly what CV measures.
