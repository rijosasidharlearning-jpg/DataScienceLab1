## Correlation

Correlation is a statistical measure that expresses the extent to which two variables are linearly related (meaning they change together at a constant rate). It's a common tool for describing simple relationships without making a statement about cause and effect. The correlation coefficient ranges from -1 to +1:

*   **+1:** Perfect positive correlation (as one variable increases, the other variable increases proportionally).
*   **-1:** Perfect negative correlation (as one variable increases, the other variable decreases proportionally).
*   **0:** No linear correlation (the variables do not move together in a linear fashion).

```python
correlation_matrix = sb1.corr()
display(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
#
```

```python
# Calculate overall means for 'age' and 'fare' from the sb1 DataFrame
mean_age = sb1['age'].mean()
mean_fare = sb1['fare'].mean()

print(f"Overall Mean Age (X̄): {mean_age:.2f}")
print(f"Overall Mean Fare (Ȳ): {mean_fare:.2f}")

# Get the values for the 890th row (index 889) for age (Xi) and fare (Yi)
row_index = 889
Xi_890 = sb1.loc[row_index, 'age']
Yi_890 = sb1.loc[row_index, 'fare']

print(f"\nAge (Xi) for 890th row (index {row_index}): {Xi_890:.2f}")
print(f"Fare (Yi) for 890th row (index {row_index}): {Yi_890:.2f}")

# Calculate the deviations for this specific row
deviation_X_890 = Xi_890 - mean_age
deviation_Y_890 = Yi_890 - mean_fare

print(f"\nDeviation (Xi - X̄) for 890th row: {deviation_X_890:.2f}")
print(f"Deviation (Yi - Ȳ) for 890th row: {deviation_Y_890:.2f}")

# The product of these deviations contributes to the numerator's sum
product_deviations_890 = deviation_X_890 * deviation_Y_890
print(f"Product of deviations for 890th row (contributes to numerator): {product_deviations_890:.2f}")
```
