### Explanation of `lsuffix` and `rsuffix`

When you join two DataFrames, if both DataFrames have columns with the same name (and those columns are *not* the keys you are joining on), pandas needs a way to differentiate them. This is where `lsuffix` (left suffix) and `rsuffix` (right suffix) come in.

-   **`lsuffix`**: A string suffix to apply to the names of overlapping columns from the *left* DataFrame.
-   **`rsuffix`**: A string suffix to apply to the names of overlapping columns from the *right* DataFrame.

In the example `pdj1.join(pdj2, lsuffix="_left", rsuffix="_right")`:

-   `pdj1` is the left DataFrame.
-   `pdj2` is the right DataFrame.

Both `pdj1` and `pdj2` have an index named 'ID'. When you perform a `join` on the index, 'ID' is the common key. However, if they had other columns with the same name (e.g., both had a 'Value' column, and you weren't joining on 'Value'), then pandas would append `_left` to `pdj1`'s 'Value' column and `_right` to `pdj2`'s 'Value' column to keep them distinct. In our `pdj1.join(pdj2)` case, `ID` is the common index, so it doesn't get a suffix. However, when we perform `pdj1.join(pdj2, lsuffix="_left", rsuffix="_right")`, `pdj1` (left) has an `ID` index and `pdj2` (right) also has an `ID` index. Since the join is performed on these indices, they become the main index of the result. If there were other columns with the same name, the suffixes would be applied to them.

Let's create a new example to clearly illustrate this with a column *other* than the index.

```python
import pandas as pd

# Create two DataFrames with a common column 'Grade' that is not the join key
df_students = pd.DataFrame({
    'StudentID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Grade': ['A', 'B', 'A']
})

df_scores = pd.DataFrame({
    'StudentID': [1, 2, 3],
    'Subject': ['Math', 'Physics', 'Chemistry'],
    'Grade': [95, 88, 92] # Another 'Grade' column
})

display(df_students)
display(df_scores)

# Now, merge them on 'StudentID'. The 'Grade' column exists in both.
# Without suffixes, pandas would raise a ValueError or automatically add suffixes.
merged_df_with_suffixes = pd.merge(df_students, df_scores, on='StudentID', suffixes=('_info', '_score'))
display(merged_df_with_suffixes)

print("Notice how 'Grade_info' comes from df_students (left) and 'Grade_score' comes from df_scores (right).")
```

### Filtering Data

You can filter the DataFrame based on conditions, for example, to see only students with 'Maths' as a subject or those with 'HasScholarship' set to 1.

```python
# Filter for students who have 'Maths' as a Subject
math_students = res1pdj1[res1pdj1['Subject'] == 'Maths']
display(math_students)

# Filter for students with 'HasScholarship' set to 1
scholarship_students = res1pdj1[res1pdj1['HasScholarship'] == 1]
display(scholarship_students)

# Filter for students with 'Physics' and Marks greater than 85
high_physics_marks = res1pdj1[(res1pdj1['Subject'] == 'Physics') & (res1pdj1['Marks'] > 85)]
display(high_physics_marks)
```

### Sorting Data

You can sort the DataFrame by one or more columns to organize your data. For example, sorting by 'Marks' in descending order.

```python
# Sort by 'Marks' in descending order
sorted_by_marks = res1pdj1.sort_values(by='Marks', ascending=False)
display(sorted_by_marks)

# Sort by 'Name' alphabetically and then by 'Marks' in descending order
sorted_by_name_and_marks = res1pdj1.sort_values(by=['Name', 'Marks'], ascending=[True, False])
display(sorted_by_name_and_marks)
```

### Further Aggregations

Beyond `sum`, you can perform other aggregations like `mean`, `count`, `min`, `max`, or custom functions. Let's look at average marks per subject or per student.

```python
# Average Marks per Subject
avg_marks_per_subject = res1pdj1.groupby('Subject')['Marks'].mean().reset_index()
display(avg_marks_per_subject)

# Count of subjects per student
subject_count_per_student = res1pdj1.groupby('Name')['Subject'].count().reset_index()
display(subject_count_per_student)

# Max Marks for each student in any subject
max_marks_per_student = res1pdj1.groupby('Name')['Marks'].max().reset_index()
display(max_marks_per_student)
```
