#Filter and SubSet

```python
#filter

fs1 = pd.Series([1,2,345,435,35,67,678,8,150,78,878,2323,2,45])

print(" greater than 100 ", fs1>100)
print(" greater than 100 and less than 200 ", fs1[(fs1>100) & (fs1<=200)])
save = fs1[fs1>100]
print(" greater than 100 ", save)
save.reset_index(drop=True)
print(" greater than 100 , without inplace or re assign ", save)
save.reset_index(drop=True, inplace=True)
print(" greater than 100 with inplace ", save)
```

```python
# Changing Index

lstdic1 = [{"a":100,"b":90},{"a":80,"b":90}]
pd2 = pd.DataFrame(lstdic1,index=["First", "Second"])
print(pd2)
```

```python
# Use .loc to access a row by its index name
print("Accessing row by label using .loc:")
print(pd2.loc["First"])

# If you wanted the column 'a', you would use:
# print(pd2["a"])
```

### Using `.iloc` for Integer-based Indexing
Unlike `.loc` which uses labels, `.iloc` uses the integer position.

```python
# Access the first row (position 0)
print("First row using iloc:")
display(pd2.iloc[0])

# Access the first element of the first row (row 0, column 0)
print("\nElement at [0,0]:", pd2.iloc[0, 0])

# Access all rows for the second column (position 1)
print("\nSecond column using iloc:")
display(pd2.iloc[:, 1])
```

##Concat

```python
pd3 = pd.DataFrame({"ID":[1,2,3,4],"Name":["A","B","C","D"]})
pd4 = pd.DataFrame({"ID":[5,6,7,8],"Name":["E","F","G","H"]})
display(pd3)
display(pd4)
```

```python
#vertical axis = 0
pd5 = pd.concat([pd3,pd4],axis=0)
display(pd5)

```

```python
# horizontal axis =1
pd6 = pd.concat([pd3,pd4],axis=1)
display(pd6)
```
